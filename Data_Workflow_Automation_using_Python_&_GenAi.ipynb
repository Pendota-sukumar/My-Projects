**Automate Data Cleaning Tasks in Python**

```python
#1. Show shape and summary
print("Initial Shape:", df.shape)
print("\nData Summary:")
print(df.info())

# 2. Remove duplicate rows
df = df.drop_duplicates()
print("\n✅ Duplicates removed.")

# 3. Handle missing values (just display for now)
print("\n🔍 Missing Values:")
print(df.isnull().sum())

# 4. Standardize column names (lowercase & no spaces)
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print("\n Standardized column names:")
print(df.columns)

# 5. Basic type check
print("\n Data Types:")
print(df.dtypes)

# 6. Outlier check for numeric fields
import numpy as np

numeric_cols = df.select_dtypes(include=np.number).columns
print("\n Outlier Detection (very high values):")
for col in numeric_cols:
    high_values = df[df[col] > df[col].mean() + 3 * df[col].std()]
    if not high_values.empty:
        print(f"Outliers in {col}:\n", high_values[[col]])
    else:
        print(f"No major outliers in {col}")
```

```python
import pandas as pd

# 1. Load your source data file (CSV for example)
file_path = "/content/OnlineRetail_CustomDataset.csv"
df = pd.read_csv(file_path)

# 2. Prepare a data summary prompt
summary = f"""
Dataset Summary:
- Total rows: {len(df)}
- Columns: {', '.join(df.columns)}
- Missing values: {df.isnull().sum().sum()}
"""

# Example: if your dataset has a 'Sales' column, add a mean sales stat
if 'Sales' in df.columns:
    summary += f"- Average Sales: {df['Sales'].mean():.2f}\n"

# Add top 3 categories by Sales if 'Category' and 'Sales' columns exist
if 'Category' in df.columns and 'Sales' in df.columns:
    top_categories = df.groupby('Category')['Sales'].sum().sort_values(ascending=False).head(3)
    summary += "- Top 3 Categories by Sales:\n"
    for cat, val in top_categories.items():
        summary += f"  * {cat}: {val:.2f}\n"

print("Prepared prompt for AI:\n", summary)

# 3. Use Together client you already initialized

response = client.chat.completions.create(
    model=model,
    messages=[
        {"role": "system", "content": "You are a helpful data analyst."},
        {"role": "user", "content": f"Generate a professional summary report based on this data summary:\n{summary}"}
    ]
)

# 4. Print the AI-generated report
print("\n📝 Auto-Generated Report:\n")
print(response.choices[0].message.content.strip())

# 5. (Optional) Save report to a file
with open("auto_generated_report.txt", "w") as f:
    f.write(response.choices[0].message.content.strip())
```
